```python
from scipy import signal
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np

path = 'C:\\Users\\silly\\Documents\\voice_pathology\\SVD_db\\'
men_pathology = 'men_pathology_all\\'
women_pathology = 'women_pathology_all\\'
men_healthy = 'men_healthy\\'
women_healthy = 'women_healthy\\'


overlap = 1024
frame_length = 2048
```

```python
from scipy.io import wavfile

def readAudio(audio):
    fs, amp = wavfile.read(audio)
    dt = 1/fs
    n = len(amp)
    t = dt*n

    if t > 1.0:
        amp = amp[int((t/2 - 0.5)/dt):int((t/2 + 0.5)/dt)]
        n = len(amp)
        t = dt*n
    
    return(amp, fs, n, t)
```

```python
amp1, fs, n ,t = readAudio(path + men_pathology + '110-a_n.wav')

fig = plt.figure(figsize=(10,4))

plt.plot(amp1)
plt.ylabel('amplitude')
plt.xlabel('msec')
plt.title('Male WITH pathology')
```

```python
amp2, fs, n ,t = readAudio(path + men_healthy + '708-a_n.wav')

fig = plt.figure(figsize=(10,4))

plt.plot(amp2)
plt.ylabel('amplitude')
plt.xlabel('msec')
plt.title('Male WITHOUT pathology')
```

```python
amp3, fs, n ,t = readAudio(path + women_pathology + '101-a_n.wav')

fig = plt.figure(figsize=(10,4))

plt.plot(amp3)
plt.ylabel('amplitude')
plt.xlabel('msec')
plt.title('Female WITH pathology')
```

```python
amp4, fs, n ,t = readAudio(path + women_healthy + '17-a_n.wav')

fig = plt.figure(figsize=(10,4))

plt.plot(amp4)
plt.ylabel('amplitude')
plt.xlabel('msec')
plt.title('Female WITHOUT pathology')
```

```python
S1 = librosa.feature.melspectrogram(y=amp1*1.0, sr=fs, n_fft=frame_length, hop_length=overlap, power=1.0)
fig = plt.figure(figsize=(10,4))

librosa.display.specshow(librosa.power_to_db(S1,ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Man WITH pathology Spectrogram')
plt.tight_layout()
```
